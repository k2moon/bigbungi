### 뉴욕 airBnB : https://www.kaggle.com/ptoscano230382/air-bnb-ny-2019 

```python
DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv'
```

```python
#데이터를 로드하고 상위 5개 컬럼을 출력하라
import pandas as pd
df = pd.read_csv(DataUrl)
df.head()
```

```python
# 데이터의 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개를 출력하라
rs1 = df.groupby('host_name').size()
rs2 = df['host_name'].value_counts().sort_index()
rs3 = df.groupby('host_name').count()
```

```python
# 데이터의 각 host_name의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터 프레임을 만들어라. 
# 빈도수 컬럼은 counts로 명명하라
rs = df.groupby('host_name').size().to_frame().rename(columns={0:'counts'}).sort_values('counts', ascending=False)
rs.head()
```

```python
# neighbourhood_group의 값에 따른 neighbourhood컬럼 값의 갯수를 구하여라
rs = df.groupby(['neighbourhood_group','neighbourhood'], as_index=False).size()
rs.head()
```

```python
# neighbourhood_group의 값에 따른 neighbourhood컬럼 값 중 neighbourhood_group그룹의 최댓값들을 출력하라
gr = df.groupby(['neighbourhood_group','neighbourhood'], as_index=False).size()
gr_max = gr.groupby('neighbourhood_group').max()
ri = gr_max.reset_index()
rs = ri[['neighbourhood_group','size']]
rs
```

```python
gr = df[['neighbourhood_group','neighbourhood']].groupby(['neighbourhood_group','neighbourhood'], as_index=False).size()
gr_max = gr.groupby('neighbourhood_group').max()
ri = gr_max.reset_index()
rs = ri[['neighbourhood_group','size']]
rs
```

```python
# neighbourhood_group 값에 따른 price값의 평균, 분산, 최대, 최소 값을 구하여라
mean = df[['neighbourhood_group','price']].groupby('neighbourhood_group').mean()
var = df[['neighbourhood_group','price']].groupby('neighbourhood_group').var()
max = df[['neighbourhood_group','price']].groupby('neighbourhood_group').max()
min = df[['neighbourhood_group','price']].groupby('neighbourhood_group').max()
rs = df[['neighbourhood_group','price']].groupby('neighbourhood_group').agg(['mean','var','max','min'])
rs
```

```python
# neighbourhood_group 값에 따른 reviews_per_month 평균, 분산, 최대, 최소 값을 구하여라
df[['neighbourhood_group','reviews_per_month']].groupby('neighbourhood_group').agg(['mean','var','max','min'])
```

```python
# neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 구하라
rs = df[['neighbourhood','neighbourhood_group','price']].groupby(['neighbourhood','neighbourhood_group']).agg('mean')
rs
```

```python
rs = df.groupby(['neighbourhood','neighbourhood_group'], as_index=False).price.agg('mean')
rs
```

```python
rs = df.groupby(['neighbourhood','neighbourhood_group']).price.agg('mean')
rs
```

```python
# neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 계층적 indexing 없이 구하라
rs = df[['neighbourhood','neighbourhood_group','price']].groupby(['neighbourhood','neighbourhood_group']).agg('mean').unstack()
rs.head()
```

```python
# neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 계층적 indexing 없이 구하고 NaN 값은 -999로 채워라
rs = df[['neighbourhood','neighbourhood_group','price']].groupby(['neighbourhood','neighbourhood_group']).agg('mean').unstack().fillna(-999)
rs.head()
```

```python
# 데이터중 neighbourhood_group 값이 Queens값을 가지는 데이터들 중 neighbourhood 그룹별로 price값의 평균, 분산, 최대, 최소값을 구하라
cond = df['neighbourhood_group'] == 'Queens'
rs = df.loc[cond].groupby('neighbourhood').price.agg(['mean','var','max','min'])
rs.head()
```

```python
#  데이터중 neighbourhood_group 값에 따른 room_type 컬럼의 숫자를 구하고 neighbourhood_group 값을 기준으로 각 값의 비율을 구하여라
gr = df.groupby(['neighbourhood_group','room_type']).size().unstack()
gr_val = gr.values
gr_sum = gr.sum(axis=1)
type(gr_val), type(gr_sum) # (numpy.ndarray, pandas.core.series.Series)
gr_sum = gr_sum.values.reshape(-1,1)
gr_cal = gr_val/gr_sum
rs = gr
rs.loc[:,:] = gr_cal
rs
```

```python
rs = gr
rs[:] = gr_cal
rs
```

### 서울시 미세먼지 데이터 : https://www.airkorea.or.kr/web/realSearch?pMENU_NO=97

```python
import pandas as pd

DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv'
df = pd.read_csv(DataUrl)
df.info()
```

```python
# 년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 
# 서울시의 제공데이터의 경우 0시가 24시로 표현된다
def func(x):
  import datetime
  date = x.split(':')[0]
  hour = x.split(':')[1]
  
  if hour == '24':
    hour = '00:00:00'    
    dt = date + ' ' + hour  
    retDT = pd.to_datetime(dt) + datetime.timedelta(days=1)
  else:
    hour = hour+':00:00'
    dt = date + ' ' + hour  
    retDT = pd.to_datetime(dt)
  return retDT

# func('2021-05-05:24')
df['(년-월-일:시)'] = df['(년-월-일:시)'].map(func)
```

```python
df.head()
```

```python
# 일자별 영어요일 이름을 dayName 컬럼에 저장하라
en_week = ['Monday','Tuesday','Wednesday','Thursday','Fiday','Saturday','Sunday']
def func(x):
  wd = x.weekday()
  return en_week[wd]

#x = pd.to_datetime('2021-05-15 15:00:00')
#func(x)

df['dayName'] = df['(년-월-일:시)'].map(func)
```

```python
dic = { 0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Fiday', 5:'Saturday', 6:'Sunday' }
df['dayName2'] = df['(년-월-일:시)'].dt.weekday
df['dayName2'] = df['dayName2'].map(dic)
```

```python
# 일자별 각 PM10등급의 빈도수를 파악하라

rs = df['PM10등급'].groupby(df['(년-월-일:시)'].dt.to_period('D')).size()
rs.head()
```
